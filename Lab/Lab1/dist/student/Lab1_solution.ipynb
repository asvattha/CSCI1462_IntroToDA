### Question 2.2 [5 points]:

The formula for computing the final amount if one is earning compound interest is given on Wikipedia as

$$
A = P (1 + \frac{r}{n})^{nt}
$$

where, 
- P = principal amount (initial investment)
- r = annual nominal interest rate (as a decimal)
- n = number of times the interest is compounded per year
- t = number of years


Write a Python program that assigns the principal amount of 10000 to variable **P**, assign to **n** the value 12, assign to **t** the value of 4, and assign to **r** the interest rate of 8% (0.08). Calculate and print the final amount after **t** years (_rounded upto two decimal points_). Your program should print out the message in the format, `The final amount after 4 years is 10269.35`

```python
P = 10000 # SOLUTION
n = 12 # SOLUTION
r = 0.08 # SOLUTION
t = 4 # SOLUTION

A = round(P * (1 + r/n)**t, 2) # SOLUTION


"""# BEGIN PROMPT
print(f"The final amount after {...} years is {...}")
"""# END PROMPT
print(f"The final amount after {t} years is {A}") # SOLUTION NO PROMPT
```

## 3. Calling a Function

### Question 3.1 [5 points]:

Take a look at the following function `repeatMsg()`, which accepts two parameters: a `str` and an `int`. The function repeats the string the specified number of times.

For example, `repeatMsg("hey", 2)` would return `heyhey`, showing that the message `hey` is repeated twice.

```python
# You may ignore this cell, since we do not know the body of the function to answer the questions. 
def repeatMsg(msg, ntimes):
    return msg * ntimes
```

```python
repeatMsg("hey", 2)
```

#### Task: 
Call the `repeatMsg` function correctly such that it returns the message, `hellohellohello`. Store this result in the variable, `your_ans`.

```python
your_ans = repeatMsg("hello", 3) # SOLUTION
print(your_ans)
```
